```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#define function that describe a distribution
def print_stats(df):
    print('Mean:', df.mean())
    print('Standard deviation:', df.std())
    print('skewness: ', df.skew())
    print('Kurtosis :', df.kurtosis())
    print('min:', df.min())
```

```python
#define a function that returns the centered and normalized data of the distribution
def cen_norm(df):
     return(df-df.mean()/df.std())
```

```python
#read distribution with read_csv function
dist = pd.read_csv('~/Downloads/distributions.csv')
dist.describe()
```

```python
#calling the print_stats function on the Normal distribution
print_stats(dist['Normal'])
```

```python
#calling the cen_norm on the distributions to centralise and normalize it 
dist['cen_normal'] = cen_norm(dist['Normal'])
dist1 = cen_norm(dist['Dist1'])
dist2 = cen_norm(dist['Dist2'])
dist3 = cen_norm(dist['Dist3'])
```

```python
#plot a histogram for the normal distribution
plt.figure()
plt.hist(dist['cen_normal'], bins = 50, range = [-4,4])
plt.show()
```
